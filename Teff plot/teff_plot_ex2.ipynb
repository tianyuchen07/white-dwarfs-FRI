```python
import numpy as np
import matplotlib.pyplot as plt
import glob
```

```python
for name in glob.glob('*.csv'):
    print(name)
```

```python
data = np.loadtxt('Teff6875logg5.344.csv', delimiter=',', skiprows=1)   

magnitude = data[:,1]
magnitudeError = data[:,2]
```

```python
def stDevCalc(magnitude, magnitudeError):
    magDev = np.std(magnitude)
    errAvg = np.average(magnitudeError)
    return magDev/errAvg
print (stDevCalc(magnitude, magnitudeError))
```

```python
def get_vals(fname):
    Teff = fname[4:8]
    logg = fname[12:17]
    Teff = float(Teff)
    logg = float(logg)
    return Teff,logg
    
# Example:
f = 'Teff8200logg4.820.csv'
Teff, logg = get_vals(f)
print(Teff,logg)
```

```python
temperature = []
variability = []
for fname in glob.glob('*.csv'):
    temperature.append(get_vals(fname)[0])
    magnitude = np.loadtxt(fname, delimiter=',', skiprows=1)[:,1]
    magnitudeError = np.loadtxt(fname, delimiter=',', skiprows=1)[:,2]
    variability.append(stDevCalc(magnitude, magnitudeError))
```

```python
plt.scatter(temperature, variability)
plt.savefig('lab2.png')
```
